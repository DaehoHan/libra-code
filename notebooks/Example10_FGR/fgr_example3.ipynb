## NE-FGR for Brownian Oscillators Model

This tutorial will demonstrate how the FGR module can be used for rate calculations

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
#import seaborn as sns
matplotlib.rcParams['savefig.dpi'] = 256
#matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['lines.linewidth'] = 2
#matplotlib.rcParams['font.family'] = 'sans-serif'

%matplotlib inline
#%matplotlib notebook
```

```python
cond = 0  #Condon=0 or non-Condon=1 
Nmethod = 5

dfa = pd.DataFrame()
for i in range(Nmethod+1):
    filename = '_res-%s-%s-0-0-0-0.txt' %  (cond, i)
    #print('Reading ',filename)
    namelist = ['t%s' % i, 'k%s' % i, 'pop%s' % i]
    df = pd.read_csv(filename, header=None, names = namelist, delim_whitespace=True)
    dfa = pd.concat([dfa, df], axis=1)
dfa.head()
```

```python
labellist = ['exact', 'LSC', 'C-AV', 'C-D', 'W-0', 'C-0']
w_da = 0
shift = -1

from matplotlib.pyplot import cm
plt.figure(1, figsize=(8, 6))
color = iter(cm.rainbow(np.linspace(0,1,Nmethod+1)))

for i in range(Nmethod+1):
    colname = 'k'+str(i)  #plot rates
    c=next(color)
    plt.plot(dfa['t0'], dfa[colname], c=c, label=labellist[i], alpha=0.9)

plt.legend(fontsize=14)
plt.title('$\omega_{DA}$=%s,s=%s' % (w_da, shift), fontsize=18)
plt.ylabel(r'$k(t)$', fontsize=18)
plt.xlabel(r't / fs', fontsize=18)
plt.tick_params(labelsize=14)
#plt.ylim(0, 1)
#plt.xlim(0, 20)
plt.show()
```

```python
labellist = ['exact', 'LSC', 'C-AV', 'C-D', 'W-0', 'C-0']
w_da = 0
shift = -1

from matplotlib.pyplot import cm
plt.figure(1, figsize=(8, 6))
color = iter(cm.rainbow(np.linspace(0,1,Nmethod+1)))
for i in range(Nmethod+1):
    colname = 'pop'+str(i)  #plot rates
    c=next(color)
    plt.plot(dfa['t0'], dfa[colname], c=c, label=labellist[i], alpha=0.9)

plt.legend(fontsize=14)
plt.title('$\omega_{DA}$=%s,s=%s' % (w_da, shift), fontsize=18)
plt.ylabel(r'Population (donor)', fontsize=18)
plt.xlabel(r't / fs', fontsize=18)
plt.tick_params(labelsize=14)
#plt.ylim(0, 1)
#plt.xlim(0, 20)
plt.show()
```
